# Seaborn

```python
import seaborn as sns
```

```python
#Basic plotting with seaborn
tip=sns.load_dataset('tips')
tip
```

```python
#Create a scatter plot 
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tip)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

```python
# Line plot
sns.lineplot(x='size', y='total_bill', data=tip)
plt.title('Line Plot of Total Bill by Size')
plt.show()
```

```python
# Categorical Plots
# Bar plot
sns.barplot(x='day', y='total_bill', data=tip)
plt.title("Bar plot of Total Bill by Day")
```

```python
# Box Plot
sns.boxenplot(x='day', y='total_bill', data=tip)
```

```python
# violin plot
sns.violinplot(x='day', y='total_bill', data=tip)
```

```python
# Histogram
sns.histplot(tip['total_bill'],bins=10,kde=True)
```

```python
# KDE plot( kernel density estimation)
sns.kdeplot(tip['total_bill'],fill=True)
```

```python
# pair plot
sns.pairplot(tip)
```

```python
# Correlation and Heatmap
# To visualize the correlation between numerical features, use a heatmap. First, select only the numerical columns and compute the correlation matrix:


corr = tip[['total_bill', 'tip', 'size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
